# Transform

```python
import ast

def array_to_sent(data):
  return " ".join(ast.literal_eval(data))
```

```python
train_preproc['question'] = train_preproc.question.apply(array_to_sent)
train_preproc['passage'] = train_preproc.passage.apply(array_to_sent)
train_preproc['seq_label'] = train_preproc.seq_label.apply(ast.literal_eval)
```

```python
def answer_to_seqlabel(passage,answer):
  token_answer = answer.split()
  token_passage = passage.split()
  res = list()
  for t in token_passage:
    if t == token_answer[0]:
      res.append('B')
    elif t in token_answer:
      res.append('I')
    else:
      res.append('O')
  return res
```

```python
from transformers import BertTokenizerFast, pipeline

tokenizer = BertTokenizerFast.from_pretrained(
    'Wikidepia/indobert-lite-squad'
)
qa_pipeline = pipeline(
    "question-answering",
    model="Wikidepia/indobert-lite-squad",
    tokenizer=tokenizer
)
```

```python
valid_preproc['question'] = valid_preproc.question.apply(array_to_sent)
valid_preproc['passage'] = valid_preproc.passage.apply(array_to_sent)
valid_preproc['seq_label'] = valid_preproc.seq_label.apply(ast.literal_eval)
```

# Predict

```python
def predict_answer(qa_pipeline,question,passage):
  result = qa_pipeline({
      'context' : passage,
      'question' : question
  })
  answer = result['answer']
  return answer_to_seqlabel(passage,answer)
```

```python
valid_preproc['predict'] = valid_preproc.apply(lambda x: predict_answer(qa_pipeline,x['question'],x['passage']),axis=1)
```

```python
submission1 = pd.DataFrame()
```

```python
submission1['Id'] = np.array(valid_preproc.index) + 1

submission1['seq_label'] = valid_preproc['predict']
```
